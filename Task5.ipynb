### `Q1:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
# write your code here
test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]

grouped = {}

for tup in test_list:
    key = tup[0]
    if key in grouped:
        grouped[key] += tup[1:]
    else:
        grouped[key] = tup
        
print(list(grouped.values()))
````

### `Q2:` Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
# write your code here
result = []
res = []
t = (1,5,7,8,10)

for i in range(len(t)):
    if i == 0:
        res = t[0] * t[1]
    elif i == (len(t) - 1):
        res = t[-1] * t[-2]
    else:
        res = t[i-1] * t[i] + t[i] * t[i+1]
    result.append(res)

print(result)
````

````python
# Same question using list comprehension

t = (1, 5, 7, 8, 10)

result = [
    t[i] * t[i+1] if i ==0
    else t[i] * t[i-1] if i == len(t) - 1
    else t[i] * t[i-1] + t[i] * t[i+1]
    for i in range(len(t))
]

print(tuple(result))
````

### `Q4`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
# write your code here
list1 = [{'hi', 'bye'}, {'Geeks', 'forGeeks'}, ('a', 'b'), ['hi', 'bye'], ['a', 'b']]

list_count = sum(1 for item in list1 if isinstance(item, list))
set_count = sum(1 for item in list1 if isinstance(item, set))
tuple_count = sum(1 for item in list1 if isinstance(item, tuple))

print(f"List-{list_count}")
print(f"Set-{set_count}")
print(f"Tuples-{tuple_count}")
````

###`Q2:` Write a program to count unique number of vowels using sets in a given string. Lowercase and upercase vowels will be taken as different.

`Input:`
```
Str1 = "hands-on data science mentorship progrAm with live classes at affordable fee only on CampusX"
```
`Output:`
```
No of unique vowels-6
```

````python
# write your code here
str1 = "hands-on data science mentorship progrAm with live classes at affordable fee only on CampusX"
s1 = set(str1)
vowels = {'a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'}
count = 0

for _ in s1:
    if _ in vowels:
        count += 1

print("Number of unique vowels - ", count)
````

### `Q1`: Key with maximum unique values

Given a dictionary with values list, extract key whose value has most unique values.

**Example 1:**

Input:

```bash
test_dict = {"CampusX" : [5, 7, 9, 4, 0], "is" : [6, 7, 4, 3, 3], "Best" : [9, 9, 6, 5, 5]}
```

Output:
```bash
CampusX
```

**Example 2:**

Input:
```bash
test_dict = {"CampusX" : [5, 7, 9, 4, 0], "is" : [6, 7, 7, 7], "Best" : [9, 9, 6, 5, 5]}
```

Output:
```bash
Best
```

````python
# write your code here
test_dict = {
    "CampusX": [5, 7, 9, 4, 0],
    "is": [6, 7, 4, 3, 3],
    "Best": [9, 9, 6, 5, 5]
}

max_key = None 
max_count = 0

for key, values in test_dict.items():
    count = len(set(values))
    if count > max_count: 
        max_count = count 
        max_key = key
    
print(max_key)
````

### `Q2`: Replace words from Dictionary. Given String, replace itâ€™s words from lookup dictionary.

**Example 1:**

Input:

```bash
test_str = 'CampusX best for DS students.'
repl_dict = {"best" : "is the best channel", "DS" : "Data-Science"}
```

Output:

```bash
CampusX is the best channel for Data-Science students.
```

**Example 2:**

Input:
```bash
test_str = 'CampusX best for DS students.'
repl_dict = {"good" : "is the best channel", "ds" : "Data-Science"}
```

Output:
```bash
CampusX best for DS students.
```

````python
# write your code here

test_str = 'CampusX best for DS students.'
repl_dict = {"best" : "is the best channel", "DS" : "Data-Science"}

words = test_str.split(" ")
result = []

for word in words:
    if word in repl_dict:
        result.append(repl_dict[word])
    else:
        result.append(word)

final_str = " ".join(result)
print(final_str)
````

````python
def convert_list_to_dict(test_list, key_list):
    chunk_size = len(key_list)
    return [dict(zip(key_list, test_list[i:i+chunk_size])) for i in range(0, len(test_list), chunk_size)]
````

````python
# Example 1
test_list1 = ["DataScience", 3, "is", 8]
key_list1 = ["name", "id"]
print(convert_list_to_dict(test_list1, key_list1))
# Output: [{'name': 'DataScience', 'id': 3}, {'name': 'is', 'id': 8}]

# Example 2
test_list2 = ["CampusX", 10]
key_list2 = ["name", "id"]
print(convert_list_to_dict(test_list2, key_list2))
# Output: [{'name': 'CampusX', 'id': 10}]

# approach second 
result = []

chunk_size = len(key_list1)
for i in range(0, len(test_list1), chunk_size):
    values = test_list1[i:i+chunk_size]
    d = dict(zip(key_list1, values))
    result.append(d)

print(result)
````

### `Q4`: Convert a list of Tuples into Dictionary.

**Example 1:**

### `Q5`: Sort Dictionary key and values List.

**Example 1:**

Input:

```bash
{'c': [3], 'b': [12, 10], 'a': [19, 4]}
```

Output:

```bash
{'a': [4, 19], 'b': [10, 12], 'c': [3]}
```

**Example 2:**

Input:

```bash
{'c': [10, 34, 3]}
```

Output:

```bash
{'c': [3, 10, 34]}
```

````python
# write your code here
# Example 1
test_dict1 = {'c': [3], 'b': [12, 10], 'a': [19, 4]}

sorted_dict1 = {key: sorted(test_dict1[key]) for key in sorted(test_dict1)}
print(sorted_dict1)
# Output: {'a': [4, 19], 'b': [10, 12], 'c': [3]}

# Example 2
test_dict2 = {'c': [10, 34, 3]}

sorted_dict2 = {key: sorted(test_dict2[key]) for key in sorted(test_dict2)}
print(sorted_dict2)
# Output: {'c': [3, 10, 34]}
````
