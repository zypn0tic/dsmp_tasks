### Q1 :- Print the given strings as per stated format.

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
x1 = int(input("Enter the first number: "))
y1 = int(input("Enter the second number: "))
x2 = int(input("Enter the third number: "))
y2 = int(input("Enter the fourth number: "))
distance = ((x2-x1)**2 + (y2-y1)**2)**0.5
print("The distance between the two points is: ", distance)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:

Output:
dogs -> 2

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# Write your code here
numerator1 = int(input("Enter the first numerator: "))
denominator1 = int(input("Enter the first denominator: "))
numerator2 = int(input("Enter the second numerator: "))
denominator2 = int(input("Enter the second denominator: "))

numerator = numerator1 * denominator2 + numerator2 * denominator1
denominator = denominator1 * denominator2

print("The sum of the two fractions is:", numerator, "/", denominator)

````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
H = int(input("Enter the height of the milk tank: "))
L = int(input("Enter the length of the milk tank: "))
B = int(input("Enter the breadth of the milk tank: "))
h = int(input("Enter the height of the glass: "))
r = int(input("Enter the radius of the glass: "))

volume_of_milk_tank = H * L * B
volume_of_glass = (3.14 * r * r * h)

number_of_glasses = volume_of_milk_tank / volume_of_glass

print("The number of glasses that can be obtained is:", number_of_glasses)
````
